```python
#importando a biblioteca do YahooFinance para baixar cotações
#Caso não tenha instalado ainda esse pacote, basta digitar !pip3 install yfinance
import yfinance as yF

# Imports para manipulação de dados
import numpy as np
import pandas as pd

# Imports para visualização
import matplotlib.pyplot as plt
import matplotlib as m
import seaborn as sns

# Imports para cálculos estatísticos
import scipy
#from scipy.stats import kurtosis, skew, shapiro
import datetime


# Imports para formatação dos gráficos
plt.style.use('fivethirtyeight')
m.rcParams['axes.labelsize'] = 14
m.rcParams['xtick.labelsize'] = 12
m.rcParams['ytick.labelsize'] = 12
m.rcParams['text.color'] = 'k'
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 20,10


#formatando valores com duas casas decimais
pd.options.display.float_format = '{:.2f}'.format
```

```python
# Carrega o dataset
#dados = pd.read_csv("dados.csv", parse_dates = True, index_col = "Data")

Cotacoes = yF.Ticker("BBDC4.SA")

# Opções de períodos 1d, 5d, 1mo, 3mo, 6mo, 1y, 2y, 5y, 10y e ytd.
dados = Cotacoes.history(period="2y")
dados.head()
```

```python
# Crie um novo DataFrame contendo os dados do período desejado
dados = dados['2022-07-01':'2023-01-01']
```

```python
# Vamos observar o nosso range de dados
print('Menor data: ', dados.index.min())
print('Maior data:', dados.index.max())
```

```python
dados1 = dados
```

```python
dados2 = dados
```

```python
dados2 = dados2.reset_index()
```

```python
dados1 = dados1.reset_index()
```

```python
date1 = dados1['Date']
fechamento1 = dados1['Close']
date2 = dados2['Date']
fechamento2 = dados2['Close']

dd1 = pd.DataFrame({'Date1': date1, 'Close1': fechamento1})
dd2 = pd.DataFrame({'Date2': date2, 'Close2': fechamento2})

data = pd.concat([dd1,dd2], axis=1)
print(data)
```

```python
data
```

```python

# Remova as informações de fuso horário das colunas de data e hora
data['Date1'] = data['Date1'].dt.tz_localize(None)
data['Date2'] = data['Date2'].dt.tz_localize(None)

# Salve o DataFrame em um arquivo Excel
data.to_excel('acao.xlsx', index=False)
```

```python
array1 = []
array2 = []
for i in range(0, len(data) - 1):
  count1 = (data['Close1'][i + 1] - data['Close1'][i]) / data['Close1'][i]
  count2 = (data['Close2'][i + 1] - data['Close2'][i]) / data['Close2'][i]

  array1.append(count1)
  array2.append(count2)

print(array1, array2)
```

```python
corre = pd.DataFrame({'Itau': array1, 'Bradesco': array2})
```

```python
correlacao = corre['Itau'].corr(corre['Bradesco'])
print(correlacao)
```
