```r
path = '/content/New_Category.csv'
data = read.csv(path)
#print(data)
```

```r
data = na.omit(data)
summary(data)
```

```r
data1 = data[, -c(1,2)]
summary(data1)
```

```r
# Matriz de correlação
correlacao <- cor(data1)
print(correlacao)
```

```r
length(data1)
# Suponha que "data1" seja o seu conjunto de dados
nomes_colunas <- colnames(data1)

# Exibir os nomes das colunas
print(nomes_colunas)
```

```r
# Crie um vetor para armazenar os resultados
resultados_anova <- vector("list", length = 94)

# Execute a ANOVA para cada coluna
for (i in 1:94) {
  # Selecione a coluna atual
  coluna_atual <- data1[, i]

  # Realize a ANOVA
  modelo_anova <- aov(Odd_Categoria ~ coluna_atual, data = data1)

  # Armazene o resultado no vetor
  resultados_anova[[i]] <- summary(modelo_anova)
}
```
