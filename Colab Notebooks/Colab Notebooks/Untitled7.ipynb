Aqui, como as questões 1 e 2 são funções polinomiais, faz mais sentido tomar apenas os coeficientes da função dada. Esse jeito facilita nosso trabalho em relação a implementação do método respectivo.

```python
from scipy.optimize import bisect

def func(x):
    return 2*(x**2) - 7*x + 1

print(func(3))
# Use full_output=True para obter informações detalhadas
root, info = bisect(func, 1, 5, full_output=True)

# Extraia o número de iterações do resultado
num_iterations = info.iterations

print("A raiz encontrada é:", root)
print("Número de iterações:", num_iterations)
```

```python
func = [input("Função a ser minimizada: ")]
a = float(input("Intervalo de Inicio: "))
b = float(input("Intervalo Final: "))
```

```python
print(type(func))
func = func[0].split(',')
print(func)
print(len(func))
```

```python
t = len(func)
k = 1
c = (a + b)/2
#c = 3
func1 = float(func[0])*(c**2) + float(func[1])*c + float(func[2])
print(func1)
```

```python
while k <= 41:
  print(3*'*-*-*-*-*-*-*-*-*-*-')
  print('Valor de k: ', k)
  c = (a + b)/2
  print("Valor de c: ", c)
  func1 = float(func[0])*(c**2) + float(func[1])*c + float(func[2])
  print("Valor de f(c_k): ", func1)
  y = (a + c)/2
  print("Valor de y_k: ", y)
  func2 = float(func[0])*(y**2) + float(func[1])*y + float(func[2])
  print("Valor de f(y_k): ", func2)
  #print(func1,func2)
  if func2 <= func1:
    print("f(y_k) <= f(c_k): ", func2, func1)
    print("a_{k+1} = a_k")
    a = a
    print("b_{k + 1} = c_k")
    b = c
    print("c_{k+1} = y_k")
    c = y
    k += 1

  else:
    print('f(y_k) > f(c_k): ', func2,func1)
    z = (c + b)/2
    print("Valor de z: ", z)
    func3 = float(func[0])*(z**2) + float(func[1])*z + float(func[2])
    print("Valor de f(z_k): ", func3)
    if func1 <= func3:
      print("f(y_k) <= f(z_k): ", func2, func3)
      print("a_{k+1} = y_k")
      a = y
      print("b_{k + 1} = z_k")
      b = z
      print("c_{k+1} = c_k")
      c = c
    else:
      print('f(y_k) > f(z_k): ', func2,func3)
      print("a_{k+1} = c_k")
      a = c
      print("b_{k + 1} = b_k")
      b = b
      print("c_{k+1} = z_k")
      c = z
    k += 1
```
