```python
data['MediaM5'] = data['odd_entrada'].rolling(window=5).mean()
```

```python
data['MediaM5'].describe()
data = data.fillna(0)
```

```python
data['MediaM5'].describe()
```

```python
t = len(data)
data['aposta'] = np.zeros(t)
data['Qt_Apostas'] = np.zeros(t)
data['Resultado'] = np.zeros(t)
data['Acerto'] = np.zeros(t)

for i in range(4, t):
  if data['MediaM5'][i] <= 5.4:
    data.loc[i,'aposta'] = 1
    data.loc[i,'Qt_Apostas'] = 1
    if data['MediaM5'][i] <= 2.6:
      data.loc[i,'Qt_Apostas'] = 2
for i in range(4,t):
  data.loc[i,'Resultado'] = data['Resultado'][i-1]
  if data['aposta'][i] == 1:
    if data['Odd_Categoria'][i] >= 5:
      data.loc[i,'Acerto'] = 1
      if data['Qt_Apostas'][i] == 1:
        data['Resultado'][i] = data['Resultado'][i] + 1
      if data['Qt_Apostas'][i] == 2:
        data['Resultado'][i] = data['Resultado'][i] + 2
    if data['Odd_Categoria'][i] < 5:
      data.loc[i,'Acerto'] = -1
      if data['Qt_Apostas'][i] == 1:
        data['Resultado'][i] = data['Resultado'][i] - 2
      if data['Qt_Apostas'][i] == 2:
        data['Resultado'][i] = data['Resultado'][i] - 4


print(data['Resultado'][t-1])
```
