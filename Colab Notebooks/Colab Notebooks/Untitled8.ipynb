```python
import pandas as pd
import numpy as np
import pulp

# Lê os dados do Excel
c1 = pd.read_excel('/content/vetorc4.xlsx', header=None).T
c1 = c1.values.flatten()

b1 = pd.read_excel('/content/vetorb4.xlsx', header=None).T
b1 = b1.values.flatten()

A1 = pd.read_excel('/content/matrizA4.xlsx', header=None)
A1 = np.array(A1)

# Cria um problema de programação linear
problema = pulp.LpProblem("Problema de Otimização", pulp.LpMinimize)

# Define as variáveis de decisão
x = [pulp.LpVariable(f'x{i}', lowBound=0) for i in range(len(c1))]

# Define a função objetivo
problema += pulp.lpSum([c * x[i] for i, c in enumerate(c1)])

# Define as restrições
for j in range(len(b1)):
    problema += pulp.lpSum([A1[j, i] * x[i] for i in range(len(c1))]) == b1[j]

# Resolve o problema
problema.solve()

# Exibe o resultado
print(f"Status da solução: {pulp.LpStatus[problema.status]}")
print(f"Valor ótimo da função objetivo: {pulp.value(problema.objective)}")
print("Valores ótimos das variáveis de decisão:")
for v in x:
    print(f"{v.name} = {v.varValue}")
```

```python
import pandas as pd
import numpy as np
import pulp

# Lê os dados do Excel
c1 = pd.read_excel('/content/vetorc4.xlsx', header=None).T
c1 = c1.values.flatten()

b1 = pd.read_excel('/content/vetorb4.xlsx', header=None).T
b1 = b1.values.flatten()

A1 = pd.read_excel('/content/matrizA4.xlsx', header=None)
A1 = np.array(A1)

u1 = pd.read_excel('/content/vetoru4.xlsx', header=None).T
u1 = u1.values.flatten()

print(c1.shape, b1.shape, A1.shape, u1.shape)

import numpy as np
from scipy.optimize import linprog

# Lê os dados do Excel e define c, A, b, e u
c = c1
A = A1
b = b1
u = u1

# Número de variáveis
n = len(c)

# Converta o problema em uma forma padrão (Ax = b, x >= 0)
# Adicione variáveis de folga para transformar as desigualdades em igualdades
A_eq = np.vstack([A, np.eye(n)])
b_eq = np.hstack([b, u])

# Resolva o problema de programação linear com o método do simplex de duas fases
result = linprog(c, A_eq=A_eq, b_eq=b_eq, method='highs')

if result.success:
    print("Solução ótima encontrada:")
    print("Valores ótimos das variáveis de decisão:")
    for i, value in enumerate(result.x):
        print(f"x{i} = {value}")
    print("Valor ótimo da função objetivo:", result.fun)
else:
    print("O problema é infeasible ou ilimitado.")

```

```python
import pandas as pd
import numpy as np
import pulp

# Lê os dados do Excel
c1 = pd.read_excel('/content/vetorc4.xlsx', header=None).T
c1 = c1.values.flatten()

b1 = pd.read_excel('/content/vetorb4.xlsx', header=None).T
b1 = b1.values.flatten()

A1 = pd.read_excel('/content/matrizA4.xlsx', header=None)
A1 = np.array(A1)

u1 = pd.read_excel('/content/vetoru4.xlsx', header=None).T
u1 = u1.values.flatten()

print(c1.shape, b1.shape, A1.shape, u1.shape)

# Cria um problema de programação linear
problema = pulp.LpProblem("Problema de Otimização", pulp.LpMinimize)

# Define as variáveis de decisão
x = [pulp.LpVariable(f'x{i}', lowBound=0) for i in range(len(c1))]

# Define a função objetivo
problema += pulp.lpSum([c * x[i] for i, c in enumerate(c1)])

# Define as restrições
for j in range(len(b1)):
    problema += pulp.lpSum([A1[j, i] * x[i] for i in range(len(c1))]) == b1[j]

# Adicione as variáveis de folga e as restrições correspondentes para o método de duas fases
z = [pulp.LpVariable(f'z{i}', lowBound=0) for i in range(len(c1))]

for i in range(len(c1)):
    problema += z[i] >= x[i]
    problema += z[i] >= -x[i]

# Resolve o problema
problema.solve()

# Exibe o resultado
print(f"Status da solução: {pulp.LpStatus[problema.status]}")
print(f"Valor ótimo da função objetivo: {pulp.value(problema.objective)}")
print("Valores ótimos das variáveis de decisão:")
for v in x:
    print(f"{v.name} = {v.varValue}")
```
