```python
import pandas as pd
import numpy as np
import re
from datetime import datetime, timedelta

path = 'bd_pseudo_1.csv'

data = pd.read_csv(path)

print('Dados carregado!')

odd = []
horario = []

t = len(data)

t0 = 0
while t0 < t:
    value = data['odd'][t0]

    if isinstance(value,str):
        aux = value.split('\n\n')
        a1 = aux[0].replace(')I','')
        a2 = a1.replace(')','')
        a3 = a2.replace('|','1')
        a4 = a3.replace('?','7')
        a5 = re.sub('[^0-9.]','',a4)
        odd.append(float(a5))
    horario.append(data['hora_criacao'][t0])
#    print(t0)
    t0 += 1

horas = []

for item in horario:
    dt = datetime.strptime(item, "%a %b %d %H:%M:%S %Y")
    horas.append(dt.hour)

data1 = pd.DataFrame({'Odd_Ajustada': odd, 'hora': horas, 'horario': horario})

print('Dados atualizados - parte 1')

tempo = []
for i in range(2, len(data1)):
    name = pd.to_datetime(data1['horario'][i - 1]) - pd.to_datetime(data1['horario'][i - 2])
    tempo.append(name)

time = np.ones(len(data1), dtype = int)
tempo_em_segundos = [t.total_seconds() for t in tempo]

for i in range(0, len(time)):
    if i == 0 or i == 1:
        time[i] = 12
    else:
        time[i] = tempo_em_segundos[i - 2]

data1['tempo_em_segundos'] = time

print('Dados atualizados - parte 2')

has_nan = data1['Odd_Ajustada'].isna().any()
if has_nan:
    print("Existem valores NaN na coluna 'Odd_Ajustada'.")
else:
    print("NÃ£o existem valores NaN na coluna 'Odd_Ajustada'.")

print("Calculo - Tempo em Segundos")
data2 = data1
data = data1

array_categoriat = []
for name in data['tempo_em_segundos']:
  if name < 12:
    array_categoriat.append(1)
  elif name >= 12 and name < 13.5:
    array_categoriat.append(2)
  elif name >= 13.5 and name < 15:
    array_categoriat.append(3)
  elif name >= 15 and name < 16.5:
    array_categoriat.append(4)
  elif name >= 16.5 and name < 19:
    array_categoriat.append(5)
  elif name >= 19 and name < 21:
    array_categoriat.append(6)
  elif name >= 21 and name < 24:
    array_categoriat.append(7)
  elif name >= 24 and name < 27:
    array_categoriat.append(8)
  elif name >= 27 and name < 32:
    array_categoriat.append(9)
  elif name >= 32 and name < 40:
    array_categoriat.append(10)
  elif name >= 40:
    array_categoriat.append(11)

array_categoriat = pd.DataFrame(array_categoriat)
array_categoriat.columns = ['tempo_segundos_categoria']
data = pd.concat([data, array_categoriat], axis = 1)

print('Coluna categoria tempo em segundos - Ok!')

array1 = []
array2 = []
array3 = []
array4 = []
array5 = []
array6 = []
array7 = []
array8 = []

#print(len(data))
inicio = len(data) - 641
print(inicio)
fim = len(data)
for i in range(inicio,fim):
    delta = fim - i
    print(i)
    #print(delta)
    if delta <= 5:
        a1 = data['tempo_segundos_categoria'][i]
        array1.append(a1)
    elif delta <= 10:
    	a2 = data['tempo_segundos_categoria'][i]
    	array2.append(a2)
    elif delta <= 20:
    	a3 = data['tempo_segundos_categoria'][i]
    	array3.append(a3)
    elif delta <= 40:
    	a4 = data['tempo_segundos_categoria'][i]
    	array4.append(a4)
    elif delta <= 80:
    	a5 = data['tempo_segundos_categoria'][i]
    	array5.append(a5)
    elif delta <= 160:
    	a6 = data['tempo_segundos_categoria'][i]
    	array6.append(a6)
    elif delta <= 320:
    	a7 = data['tempo_segundos_categoria'][i]
    	array7.append(a7)
    elif delta <= 640:
    	a8 = data['tempo_segundos_categoria'][i]
    	array8.append(a8)

print(len(array1))
#b1 = sum(array1)
#print(b1)
print(len(array2))
#b2 = sum(array2)
#print(b2)
print(len(array3))
#b3 = sum(array3)
#print(b3)
print(len(array4))
#b4 = sum(array4)
#print(b4)
print(len(array5))
#b5 = sum(array5)
#print(b5)
print(len(array6))
#b6 = sum(array6)
#print(b6)
print(len(array7))
#b7 = sum(array7)
#print(b7)
print(len(array8))
#b8 = sum(array8)
#print(b8)

#print(array4)
#print(array5)
#print(array6)
#print(array7)
#print(array8)
```
