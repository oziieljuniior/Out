```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/home/darkcover/Documentos/Out/dados/odds_200k.csv")
data.head()
```

```python
array1 = []
array2 = []
medias = []
for name in data['Odd']:
    if name >= 1.75:
        array1.append(1)
    else:
        array1.append(0)

for i in range(len(array1)):
    array2.append(array1[i])
    if i % 60 == 0 and i != 0:
        media = sum(array2) / 60
        medias.append(media)
        array2 = []


print(len(array2))
print(medias)
print(len(medias))
```
