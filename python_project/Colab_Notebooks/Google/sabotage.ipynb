```python
# Dividir os dados em treinamento e teste
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.2, random_state=42)

# Criar o modelo LSTM com camadas adicionais e dropout
model1 = Sequential()
model1.add(LSTM(64, input_shape=(X_train1.shape[1], 1), return_sequences=True))  # Camada LSTM com retorno de sequÃªncia
model1.add(Dropout(0.2))  # Camada de dropout para reduÃ§Ã£o de overfitting
model1.add(LSTM(64))  # Segunda camada LSTM
model1.add(Dense(64, activation='relu', kernel_regularizer=l2(0.001)))  # Camada densa com regularizaÃ§Ã£o L2
model1.add(Dense(2, activation='softmax'))  # Camada de saÃ­da com ativaÃ§Ã£o softmax para 11 categorias

# Compilar o modelo com uma taxa de aprendizado menor
model1.compile(loss='sparse_categorical_crossentropy', optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), metrics=['accuracy'])

# Treinar o modelo
model_train1 = model1.fit(X_train1.reshape(X_train1.shape[0], X_train1.shape[1], 1), y_train1, epochs=10, batch_size=64, validation_split=0.2)
#model_train

# Avaliar o modelo no conjunto de teste
accuracy = model1.evaluate(X_test1.reshape(X_test1.shape[0], X_test1.shape[1], 1), y_test1)[1]
print(f'AcurÃ¡cia do modelo: {accuracy}')

# Fazer previsÃµes
predictions = model1.predict(X_test1.reshape(X_test1.shape[0], X_test1.shape[1], 1))

# Arredondar as previsÃµes para obter rÃ³tulos de classe
predicted_labels = np.argmax(predictions, axis=1)

# Calcular o F1-Score
f1 = f1_score(y_test1, predicted_labels, average='weighted')

print(f'F1-Score do modelo: {f1}')
```

```python
from keras.utils import to_categorical

y2 = to_categorical(y1, num_classes=2)

# Dividir os dados em treinamento e teste
X_train2, X_test2, y_train2, y_test2 = train_test_split(X1, y2, test_size=0.2, random_state=42)

# Criar o modelo LSTM com camadas adicionais e dropout
model2 = Sequential()
model2.add(LSTM(256, input_shape=(X_train2.shape[1], 1), return_sequences=True))  # Camada LSTM com retorno de sequÃªncia
model2.add(Dropout(0.2))  # Camada de dropout para reduÃ§Ã£o de overfitting
model2.add(LSTM(256))  # Segunda camada LSTM
model2.add(Dense(256, activation='relu', kernel_regularizer=l2(0.001)))  # Camada densa com regularizaÃ§Ã£o L2
model2.add(Dense(2, activation='softmax'))  # Camada de saÃ­da com ativaÃ§Ã£o softmax para 2 categorias

# Compilar o modelo com uma taxa de aprendizado menor
model2.compile(loss='categorical_crossentropy', optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), metrics=['accuracy'])

# Treinar o modelo
model_train2 = model2.fit(X_train2.reshape(X_train2.shape[0], X_train2.shape[1], 1), y_train2, epochs=5, batch_size=256, validation_split=0.2)
#model_train

# Avaliar o modelo no conjunto de teste
accuracy1 = model2.evaluate(X_test2.reshape(X_test2.shape[0], X_test2.shape[1], 1), y_test2)[1]
print(f'AcurÃ¡cia do modelo: {accuracy1}')

# Fazer previsÃµes
predictions1 = model2.predict(X_test2.reshape(X_test2.shape[0], X_test2.shape[1], 1))
```

```python
# Aplicar um limiar (threshold) de 0.5 para obter rÃ³tulos de classe binÃ¡rios
predicted_labels_binary = (predictions1[:, 1] > 0.5).astype(int)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Calcular mÃ©tricas de avaliaÃ§Ã£o
accuracy_binary = accuracy_score(y_test2[:, 1], predicted_labels_binary)
precision_binary = precision_score(y_test2[:, 1], predicted_labels_binary)
recall_binary = recall_score(y_test2[:, 1], predicted_labels_binary)
f1_binary = f1_score(y_test2[:, 1], predicted_labels_binary)

print(f'AcurÃ¡cia: {accuracy_binary}')
print(f'PrecisÃ£o: {precision_binary}')
print(f'Recall: {recall_binary}')
print(f'F1-Score: {f1_binary}')
```

```python
from keras.optimizers import Nadam
from keras.utils import to_categorical

y3 = to_categorical(data['Odd_Categoria2'].values, num_classes=2)
# Dividir os dados em treinamento e teste
X_train3, X_test3, y_train3, y_test3 = train_test_split(X1, y3, test_size=0.2, random_state=42)

# Criar o modelo LSTM com camadas adicionais e dropout
model3 = Sequential()
model3.add(LSTM(64, input_shape=(X_train3.shape[1], 1), return_sequences=True))  # Camada LSTM com retorno de sequÃªncia
model3.add(Dropout(0.2))  # Camada de dropout para reduÃ§Ã£o de overfitting
model3.add(LSTM(64))
model3.add(Dense(2, activation='relu',kernel_regularizer=l2(0.001)))  # Camada densa com regularizaÃ§Ã£o L2
model3.add(Dense(2, activation='softmax'))  # Camada de saÃ­da com ativaÃ§Ã£o softmax para 11 categorias

# Compilar o modelo com uma taxa de aprendizado menor  #tf.keras.optimizers.Adam(learning_rate=0.001)
model3.compile(loss='categorical_crossentropy', optimizer= Nadam(learning_rate=0.002, beta_1=0.9, beta_2=0.999), metrics=['accuracy'])

# Treinar o modelo
model_train3 = model3.fit(X_train3.reshape(X_train3.shape[0], X_train3.shape[1], 1), y_train3, epochs=10, batch_size=128, validation_split=0.25)
#model_train

# Avaliar o modelo no conjunto de teste
accuracy2 = model3.evaluate(X_test3.reshape(X_test3.shape[0], X_test3.shape[1], 1), y_test3)[1]
print(f'AcurÃ¡cia do modelo: {accuracy2}')

# Fazer previsÃµes
predictions2 = model3.predict(X_test3.reshape(X_test3.shape[0], X_test3.shape[1], 1))

# Arredondar as previsÃµes para obter rÃ³tulos de classe
predicted_labels2 = np.argmax(predictions2, axis=1)
```
