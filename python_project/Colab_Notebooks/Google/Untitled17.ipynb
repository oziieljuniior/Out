```python
dados_puro=np.copy(x)
y=np.copy(y_new)
```

```python
x_train, x_test, y_train, y_test = train_test_split(dados_puro, y, test_size=0.2, random_state=42)
```

```python
# Model / data parameters
num_classes = 2
input_shape = (64, 1, 1)

# Load the data and split it between train and test sets
# (x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Scale images to the [0, 1] range
x_train = x_train.astype("float32") / 571831.230000
x_test = x_test.astype("float32") / 571831.230000
# Make sure images have shape (28, 28, 1)
x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test, -1)
print("x_train shape:", x_train.shape)
print(x_train.shape[0], "train samples")
print(x_test.shape[0], "test samples")


# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
model = keras.Sequential(
    [
        keras.Input(shape=input_shape),
        #layers.Conv2D(32, kernel_size=(3, 3), activation="relu"),
        #layers.MaxPooling2D(pool_size=(2, 2)),
        #layers.Conv2D(64, kernel_size=(3, 3), activation="relu"),
        #layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Flatten(),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation="relu"),#funcao personalizada
        layers.Dense(num_classes, activation="softmax"),#camada densa softmax
    ]
)

model.summary()
```

```python
from keras.optimizers import Nadam
batch_size = 128
epochs = 10
nadam = Nadam(learning_rate=0.002, beta_1=0.9, beta_2=0.999)
model.compile(optimizer=nadam, loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.02)
```

```python
score = model.evaluate(x_test, y_test, verbose=0)
print("Loss:", score[0])
print("Accuracy:", score[1])
```
